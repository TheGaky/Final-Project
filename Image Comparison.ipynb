```python
dick_files = []
for i in main:
    for j in minor:
        im1, im2 = main_folder+"/"+i, minor_folder+"/"+j
        
        #getting pixel values
        img1 = cv2.imread(im1)
        img2 = cv2.imread(im2)

        #converting images to grayscale
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
        
        dim = (300, 500)
        
        resized1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
        resized2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)
        
        h, w = img1.shape
        diff = cv2.subtract(resized1, resized2)
        err = np.sum(diff**2)
        mse = err/(float(h*w))
        if mse == 0:
            os.remove(im2)
            minor.remove(j)
            print(im2)
```

```python
def compare_folders(folder1, folder2):
    
    images_paths1 = os.listdir(folder1)
    images_paths2 = os.listdir(folder2)
    
    
    for i in images_paths1:
        if i == ".DS_Store":
            images_paths1.remove(i)
        
    for i in images_paths2:
        if i == ".DS_Store":
            images_paths2.remove(i)
            
    #size of the biggest folder
    images1 = np.zeros((len(images_paths1), 300, 500, 3))
    
    images2 = np.zeros((len(images_paths1), 300, 500, 3))
    
    for i in range(len(images_paths1)):
        photo = folder1 + "/" + images_paths1[i]
        photo = cv2.imread(photo)
        photo = cv2.resize(photo, (500, 300))
        images1[i] = photo
    images1 = images1.reshape(300, -1, 3, 500)
        
    for i in range(len(images_paths2)):
        photo = folder2 + "/" + images_paths2[i]
        photo = cv2.imread(photo)
        photo = cv2.resize(photo, (500, 300))
        images2[i] = photo
    images2 = images2.reshape(300, -1, 3, 500)

    for i in range(len(images_paths1)):
        #print(images2.shape)
        image1 = images1[100:200][i]

        result = images2[100:200] - image1
        #shit.append(result) 
        #print(images1.shape)
        #print(images2.shape)
        #print(result.shape)
        for j in result:
            if np.abs(np.sum(j)) < 0.5:
                print(i, np.sum(j))
```
